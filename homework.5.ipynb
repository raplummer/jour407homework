```python
import agate
```

```python
unlcrimes = agate.Table.from_csv('Downloads/unlcrime2.csv')
```

I needed to load my data as a table

```python
print(unlcrimes)

```

I needed to print the table to see the data

```python
codes = unlcrimes.group_by('Building')
code_counts = codes.aggregate([
    ('count', agate.Length())
])
code_counts = code_counts.order_by('Building')
```

I wanted a specific type of data. I used the group by formula and I wanted a count of that data

```python
code_counts.print_table()
```

I printed the table that would show me the data i was looking for. I now have the buildings were the crime occured and the number of counts. This table shows a large number of incomplete data.
